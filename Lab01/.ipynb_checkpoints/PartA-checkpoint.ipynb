```python
import nltk
import re
```

```python
nltk.download('wordnet')
```

```python
nltk.download('stopwords')
from nltk.corpus import stopwords
english_stopwords = stopwords.words("English")
```

```python
from nltk.stem import PorterStemmer
from nltk.tokenize import word_tokenize
wn = nltk.WordNetLemmatizer()
ps = nltk.PorterStemmer()
```

```python
text = '''
I am a fourth-year bachelor's student in Data Science and AI, an exciting field that combines
cutting-edge technology with the power of data to solve real-world problems! 
'''
```

```python
import string

def remove_punctuation(text):
    return "".join([char for char in text if char not in string.punctuation])
```

```python
def performStemming(text):
    return" ".join([ps.stem(word) for word in re.split('\W+', text)])
```

```python
def performLemmatization(text):
    return" ".join([wn.lemmatize(word,'v') for word in re.split('\W+', text)])
```

```python
def tokenize(text):
    return nltk.word_tokenize(text)
```

```python
def clean_tokens(tokenized_words):
    filtered_words = []
    for word in tokenized_words:
        if word not in english_stopwords:
            filtered_words.append(word)

    return filtered_words
```

```python
def preprocess(text):
    text = remove_punctuation(text)
    text_tokenized = tokenize(text)
    tokenized_clean = clean_tokens(text_tokenized)
    print(tokenized_clean)
    tokenized_clean_text = " ".join(tokenized_clean)
    stemed= performStemming(tokenized_clean_text)
    lemmatized= performLemmatization(tokenized_clean_text)
    return stemed, lemmatized
    
    
```

```python
stemed, lemmatized = preprocess(text)
```
